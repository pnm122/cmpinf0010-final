```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
arrest_data = pd.read_csv("arrest-data.csv")
census_data = pd.read_excel("census-data.xlsx")
```

```python
arrest_data["INCIDENTNEIGHBORHOOD"].value_counts()
```

```python
arrest_data.groupby("OFFENSES")["INCIDENTNEIGHBORHOOD"].head()
```

```python
crimes = {}
count = 0
weight = 0
#have weight increase based be dependant on crime commited
for arrest in arrest_data.ARRESTTIME:
    weight = 1
    
    if " 3502" in str(arrest_data["OFFENSES"][count]):
        #burgalry
        weight = weight + 2
    if " 2701" in str(arrest_data["OFFENSES"][count]):
        #assault
        weight = weight + 3
    if " 2702" in str(arrest_data["OFFENSES"][count]):
        #aggrevated assault
        weight = weight + 4
    if " 4304" in str(arrest_data["OFFENSES"][count]):
        #endangering kids
        weight = weight + 2
    if " 3304" in str(arrest_data["OFFENSES"][count]):
        #criminal mischief
        weight = weight + 2
    if " 903" in str(arrest_data["OFFENSES"][count]):
        #criminal conspiracy
        weight = weight + 3
    if " 901" in str(arrest_data["OFFENSES"][count]):
        #crime attempt
        weight = weight + 2
    if " 2901" in str(arrest_data["OFFENSES"][count]):
        #kidnapping
        weight = weight + 7
    if arrest_data["INCIDENTNEIGHBORHOOD"][count] in crimes.keys():
        crimes[arrest_data["INCIDENTNEIGHBORHOOD"][count]] = crimes[arrest_data["INCIDENTNEIGHBORHOOD"][count]] + weight
    else:
        crimes[arrest_data["INCIDENTNEIGHBORHOOD"][count]] = weight

    count = count + 1
```

```python
print(crimes)
```

```python
crimesdf = pd.DataFrame.from_dict(crimes, orient='index')
```

```python
crimesdf
```

```python
crimes2 = {}
count = 0
for arrest in arrest_data.ARRESTTIME:
    offenses = str(arrest_data["OFFENSES"][count]).split("/")
    for offense in offenses:
        if not arrest_data["INCIDENTNEIGHBORHOOD"][count] in crimes2.keys():
            crimes2[str(arrest_data["INCIDENTNEIGHBORHOOD"][count])] = {}
        if offense in crimes2[str(arrest_data["INCIDENTNEIGHBORHOOD"][count])].keys():
            crimes2[str(arrest_data["INCIDENTNEIGHBORHOOD"][count])][offense] = crimes2[str(arrest_data["INCIDENTNEIGHBORHOOD"][count])][offense] + 1
        else:
            crimes2[str(arrest_data["INCIDENTNEIGHBORHOOD"][count])][offense] = 1
    count = count + 1
```

```python
print(crimes2)
```

```python
crimes2df = pd.DataFrame.from_dict(crimes2, orient='index')
```
